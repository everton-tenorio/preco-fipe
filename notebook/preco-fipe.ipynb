```python
import requests
import plotly.graph_objects as go

url_api = 'https://veiculos.fipe.org.br/api/veiculos'

url_tabelaref = f'{url_api}/ConsultarTabelaDeReferencia'
url_marcas = f'{url_api}/ConsultarMarcas'
url_modelos = f'{url_api}/ConsultarModelos'
url_ano_modelos = f'{url_api}/ConsultarAnoModelo'
url_todos_parametros = f'{url_api}/ConsultarValorComTodosParametros'

def req(url, data):
  return requests.post(url, data=data).json()

req(url_tabelaref,'')
```

```python
dados = {
    'codigoTipoVeiculo': 2,
    'codigoMarca': 80,
    'codigoModelo': 3963,
    'ano': '2010-1',
    'anoModelo': 2010,
    'codigoTipoCombustivel': 1,
    'tipoConsulta': 'tradicional',
    'codigoTabelaReferencia': 309
}

req(url_todos_parametros, dados)


# lista de referências (meses)
referencias = req(url_tabelaref, '')
referencias = referencias[0:5][::-1] # meses em ordem crescente

# Listas para armazenar os dados do gráfico
meses = []
precos = []

# Iterando sobre as referências e obtendo os preços
for referencia in referencias:
    codigo_tabela_referencia = referencia['Codigo']
    dados['codigoTabelaReferencia'] = codigo_tabela_referencia
    response = req(url_todos_parametros, dados)
    preco_str = response['Valor'].replace('R$ ', '').replace('.', '').replace(',', '.')
    preco = float(preco_str)
    meses.append(referencia['Mes'])
    precos.append(preco)

fig = go.Figure(data=go.Scatter(x=meses, y=precos, mode='lines+markers'))

fig.update_layout(
    title='Preço Médio por Mês',
    xaxis_title='Mês',
    yaxis_title='Preço Médio (R$)',
    xaxis=dict(type='category'),  # Definindo o eixo X como categórico
    margin=dict(t=50, b=10, l=20, r=20),
    width=500,
)

fig.show()
```
